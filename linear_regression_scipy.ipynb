# Linear Regression

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt


page_speed = np.random.normal(3.0 , 1.0 , 1000)
purchase_speed = 100 - (page_speed + np.random.normal(0 , 0.1 , 1000)) * 3
plt.scatter(page_speed,purchase_speed)
plt.show()
```

```python
slope , intercept , r_value , p_value , std_err = stats.linregress(page_speed,purchase_speed)
print(r_value**2)
```

```python
def predict(x):
    return slope * x + intercept
prediction = predict(page_speed)
plt.scatter(page_speed,purchase_speed)
plt.plot(page_speed, prediction , c = 'r')
plt.show()
```

# Polynomial Regression

```python
from sklearn.metrics import r2_score
np.random.seed(2)
page_speed = np.random.normal(3.0 , 1.0 , 1000)
purchase_speed = np.random.normal(50.0,10.0,1000)/page_speed
plt.scatter(page_speed,purchase_speed)
plt.show()
```

```python
x = np.asarray(page_speed)
y = np.asarray(purchase_speed)

p4 = np.poly1d(np.polyfit(x,y,4)) #using polyfit function in numpy to fit order 4 polynomial
```

```python
plt.scatter(x,y)
xp = np.linspace(0,7 , 100)
plt.plot(xp , p4(xp) , c = 'r')
plt.show()
```

```python
print('fitting score : ',r2_score(y,p4(x)))
```
